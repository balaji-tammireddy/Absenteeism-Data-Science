##### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
from sklearn import preprocessing,linear_model
```

##### Load Data

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv_data.copy()
df
```

##### Data Preprocessing

```python
df.info()
```

##### Drop ID as they are not useful for the model

```python
df = df.drop(['ID'],axis=1)
```

```python
df
```

##### Refactoring Reason of Absence column

```python
df['Reason for Absence']
```

##### Creting dummies

```python
reason_columns = pd.get_dummies(df['Reason for Absence'],drop_first=True)
```

```python
reason_columns
```

```python
reason_columns['Check'] = reason_columns.sum(axis=1)
```

```python
reason_columns
```

```python
reason_columns['Check'].sum(axis=0)
```

```python
reason_columns['Check'].unique()
```

```python
reason_columns = reason_columns.drop(['Check'], axis = 1)
```

```python
reason_columns
```

##### Dropping Reason for Absence column

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
df
```

```python
reason_columns.columns.values
```

##### Grouping Reasons

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
```

##### Adding Reason groups to the DataFrame

```python
df = pd.concat([df, reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
```

```python
df
```

##### Renaming column names

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2','Reason_3','Reason_4']
```

```python
df.columns = column_names
```

```python
df
```

##### Reordering Column Names

```python
column_names_reordered = [ 'Reason_1', 'Reason_2','Reason_3','Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

##### Creating Checkpoint

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod
```

##### Reformatting Date Column

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
```

```python
df_reason_mod['Date']
```

##### Extracting Month Column

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
for i in range(700):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod
```

##### Extracting Dates

```python
df_reason_mod['Date'][699].weekday()
```

```python
def date_to_week(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_week)
```

```python
df_reason_mod.head()
```

##### Removing Date column and reordering the columns

```python
df_reason_mod = df_reason_mod.drop(['Date'],axis=1)
```

```python
df_reason_mod
```

```python
df_reason_mod.columns.values
```

```python
col = ['Reason_1','Reason_2','Reason_3','Reason_4','Month Value','Day of the Week','Transportation Expense','Distance to Work','Age','Daily Work Load Average','Body Mass Index','Education','Children','Pets','Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[col]
```

```python
df_reason_mod
```

##### Creating Checkpoint

```python
df_reason_date_mod = df_reason_mod.copy()
```

##### Preprocessing Education column

```python
df_reason_date_mod['Education'].value_counts()
```

```python
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_reason_date_mod
```

```python
df_reason_date_mod['Education'].value_counts()
```
